```python
class Tree(object):
    """
    A Tree class where root can be any object, just like 
    a list. We'll most likely have it as char or a number.
    We'll initiate the tree with no left/right children, but
    when needed they will also be represented by the Tree class.
    Making them sub trees.
    """
    
    def __init__(self, root):
        self.key = root
        self.leftChild = None
        self.rightChild = None
    
    def insert_left_child( self, node ):
        # function to insert a left child to a 
        # given node.
        if self.leftChild == None:
            # If no left child is present insert it
            self.leftChild = Tree(node)
        else:
            # If there is a left child already present, make the
            # new one the left child and append the existing left 
            # child to the new left child.
            temp = Tree(node)
            temp.leftChild = self.leftChild
            self.leftChild = temp
            
    def insert_right_child( self, node ):
        # function to insert a right child to a 
        # given node. Similar to insert_left_child
        # function defined above.
        if self.rightChild == None:
            self.rightChild = Tree(node)
        else:
            temp = Tree(node)
            temp.rightChild = self.rightChild
            self.rightChild = temp

    def preorder_traversal( self ):
        # Function to do a preorder traversal of the 
        # tree, root-->leftchild-->rightchild
        print self.key
        if self.leftChild is not None:
            self.leftChild.preorder_traversal()
        if self.rightChild is not None:
            self.rightChild.preorder_traversal()
            
    def postorder_traversal( self ):
        # Function to do a postorder traversal of the 
        # tree, leftchild-->rightchild-->root
        if self.leftChild is not None:
            self.leftChild.postorder_traversal()
        if self.rightChild is not None:
            self.rightChild.postorder_traversal()
        print self.key
        
    def inorder_traversal( self ):
        # Function to do a postorder traversal of the 
        # tree, leftchild-->rightchild-->root
        if self.leftChild is not None:
            self.leftChild.postorder_traversal()
        print self.key
        if self.rightChild is not None:
            self.rightChild.postorder_traversal()
            
# Function to build the tree for testing purposes
# with the structure:
#      1 
#     / \ 
#   2    3 
#  /\   /\
# 4 5  6 7
def build_test_tree():
    testTree = Tree(1)
    testTree.insert_left_child(2)
    testTree.insert_right_child(3)
    testTree.leftChild.insert_left_child(4)
    testTree.leftChild.insert_right_child(5)
    testTree.rightChild.insert_left_child(6)
    testTree.rightChild.insert_right_child(7)
    return testTree
```

> Iterative pre-order, post-order and in-order tree traversals

> Source : Nothing in particular, heard from friends. Some one was asked to do this in an interview. A good ref however would be wiki article - http://en.wikipedia.org/wiki/Tree_traversal

```python
def iterative_pre_order( root ):
    # A pre-order traversal is root->leftChild->rightChild.
    # It is pretty elegant and easy with recursion. In the 
    # iterative method, this becomes a simple depth first search
    # we start a todo list with the root (as current node), add the 
    # right child and then the left child of the current node to
    # the todo list(given they are not None), pop the last element 
    # of the to do list and make it the current node and keep going
    # until we check off every element in the todo list. The only thing
    # to note here is that we add the rightChild first and then the 
    # leftChild, making the leftChild the next node to visit.
    todoList = [root]
    preOrder = []
    while todoList:
        currNode = todoList.pop(-1)
        preOrder.append( currNode.key )
        if currNode.rightChild is not None:
            todoList.append( currNode.rightChild )
        if currNode.leftChild is not None:
            todoList.append( currNode.leftChild )
    return preOrder

def iterative_in_order( root ):
    # An inorder traversal is leftChild->root->rightChild.
    # The iterative method employed is a slightly modified
    # version of the DFS. Since we have to start at the leftmost
    # node keep adding left nodes to the todo list from the root,
    # until we hit None. Then add the last left node to the inorder
    # list. Now start moving towards the rightChild and add them 
    # to the todo list. Repeat the process, once done pop the last
    # element of the list and repeat. Simply put we keep adding
    # leftChildren from root to the todo list until no more are  
    # left and then get to the rightChild do the same and get to root.
    todoList = []
    currNode = root
    inOrderList = []
    while ( (currNode is not None) or (len(todoList) > 0) ):
        if currNode is not None:
            todoList.append( currNode )
            currNode = currNode.leftChild
        else:
            currNode = todoList.pop(-1)
            inOrderList.append( currNode.key )
            currNode = currNode.rightChild
    return inOrderList

def iterative_post_order( root ):
    # An postorder traversal is leftChild->rightChild->root.
    # iterative post order is similar to iterative in order.
    # Except instead of adding the root node to the visited/postorder
    # list, we check for right children and then go to the rightChild
    # and repeat the process. There are a couple of corner cases which
    # need to be avoided. 
    todoList = []
    currNode = root
    postOrder = []
    lastNode = None
    peek = None
    while ( (len(todoList) > 0) or (currNode is not None) ):
        if currNode is not None:
            todoList.append( currNode )
            currNode = currNode.leftChild
        else:
            # Here we'll check if the latest element of the todoList
            # if it has a right child and we didn't visit it previosly
            # we'll add it to the todoList, else add it to postOrder.
            peek = todoList[-1]
            if ( (peek.rightChild is not None) and (lastNode != peek.rightChild) ):
                currNode = peek.rightChild
            else:
                todoList.pop(-1)
                postOrder.append( peek.key )
                lastNode = peek                
    return postOrder

testTree = build_test_tree()
print "#### Iterative pre-order ####"
print iterative_pre_order(testTree)
print "#### Iterative pre-order ####"
print "#### Iterative in-order ####"
print iterative_in_order(testTree)
print "#### Iterative in-order ####"
print "#### Iterative post-order ####"
print iterative_post_order(testTree)
print "#### Iterative post-order ####"
```

> Find the maximum sum leaf to root path in a binary tree

```python
def max_root_leaf_path(root):
    # The problem can be approached in 2 steps
    # 1) Find the leaf node which results in max sum
    # 2) Find the path between the given leaf and the root.
    # We'd generally set maxNow to some really low value
    # but this i feel is more elegant way, where we change
    # maxNow to currMax if it is None.
    maxInput = None
    # we'll use a helper function to get the max leaf Node
    (maxSum, maxLeafNode) = get_max_leaf( root, maxInput, currMax = 0, maxLeaf=root )
    # Now get the max path from the max leaf
    maxPath = get_max_path( root, maxLeafNode )
    return maxSum, maxPath
    
def get_max_leaf( node, maxNow, currMax=0, maxLeaf=None ):
    # This will be a recursive function, so
    # get the base case first.
    if node is None:
        return maxNow, maxLeaf
    if maxNow is None:
        maxNow = node.key
        currMax = node.key
    else:
        currMax = currMax + node.key
    if currMax >= maxNow :
        maxNow = currMax
        maxLeaf = node
    # recurse over the left and right sub trees
    (maxNow, maxLeaf) = get_max_leaf( node.leftChild, maxNow, currMax=currMax, maxLeaf=maxLeaf )
    (maxNow, maxLeaf) = get_max_leaf( node.rightChild, maxNow, currMax=currMax, maxLeaf=maxLeaf )
    return (maxNow, maxLeaf)

def get_max_path( root, leaf, path=[] ):
    # get the base case first
    if root is None:
        return False
    # Now return true if the current node is leaf 
    # or if the path recursively leads to the leaf
    # from either left or right child
    if ( (root == leaf) or \
        (get_max_path(root.leftChild,leaf, path=path)) or\
        (get_max_path(root.rightChild,leaf, path=path)) ):
        path.append( root.key )
        return path
    return False
    
# Build the tree for testing
testTree = build_test_tree()
# get max sum path
print "#### Max Sum leaf to root path ####"
maxSum,maxPath = max_root_leaf_path(testTree)
print 'max sum-->', maxSum, ', max leaf to root Path-->', maxPath
print "#### Max Sum leaf to root path ####"
```

> Given a node and a certain distance as inputs, from a tree (root given as input), get all the nodes that the given distance apart from the input node.

> Source : http://www.geeksforgeeks.org/print-nodes-distance-k-given-node-binary-tree/

```python
def get_dist_nodes( node, root, dist ):
    # This problem is much simpler than it sounds.
    # We'll divide this problem into two parts 
    # 1) get all nodes below the input node (sub tree)
    # at the given dist from the input node.
    # 2) get all nodes that we can traverse through the
    # parent of the given node which are at the given dist
    # we'll call the two functions seperately
    nodes_dist = []
    # A stupid corner case, if dist is 0, return the same node.
    if dist == 0 :
        return [node.key]
    # get nodes from the subtree formed by input node
    nodes_dist += sub_nodes( node, dist )
    # get the route from root to the node.
    parentNodes = get_root_node_path( root, node )
    # Now get the possible paths from parent nodes
    # remove the inputNode from parent list
    parentNodes = parentNodes[1:]
    upperNodes = []
    # We simlpy keeping moving towards the root
    # by reducing the dist by 1 until either the
    # dist becomes 0 or we reach the parent node
    parentDist = dist
    while (len(parentNodes)>0) :
        parentDist -= 1
        if parentDist<0 :
            break
        pNode = parentNodes.pop(0)
        # The traverse_up func is pretty similar to 
        # the sub_nodes function, infact with very small
        # changes 1 function can be used for same. I'm writing 
        # this for ease in understanding and to be honest I 
        # realized this after coding it up.
        outPNode = traverse_up( pNode, node, parentDist )
        if outPNode is not None:
            upperNodes += outPNode
    nodes_dist += upperNodes
    return nodes_dist

def sub_nodes( inpNode, dist, count=0, nodeList = [] ):
    # A simple modified version of preorder traversal
    # just keep recursively traversing until we hit 
    # nodes at input distance or hit None and return.
    count += 1
    # This is a recursive function, so lets get the base case first
    # we are doing a count+1 in the first step so we'll check for 
    # count > dist than count == dist.
    if ( (inpNode is None) or (count > dist) ):
        # add the node to list if count > 2
        if inpNode is not None:
            nodeList.append( inpNode.key )
        return
    sub_nodes( inpNode.leftChild, dist, count=count, nodeList=nodeList )
    sub_nodes( inpNode.rightChild, dist, count=count, nodeList=nodeList )
    return nodeList

def get_root_node_path( parent, node, path=[] ):
    # A helper function which is used in the main function
    # to get all the parent nodes of input node
    # again since this is a recursive function, get the base case
    if parent is None :
        return
    if ( (parent == node) or \
        ( get_root_node_path(parent.leftChild,node) ) or \
        ( get_root_node_path(parent.rightChild,node) ) ):
        path.append(parent)
        return path
    
def traverse_up( currNode, inpNode, dist, count=0, outNodes=[], visited = [] ):
    # Now we'll run into having zero dist, 
    # take that into account.
    count += 1
    if dist == 0:
        return [currNode.key]
    # now we already took care of sub tree from input node
    # So exit when we hit input node.
    if currNode == inpNode:
        return
    # Another basecase
    if currNode == None:
        return
    if count > dist:
        if currNode not in visited:
            outNodes.append(currNode.key)
        return
    # add nodes to visited list, so we dont have to visit them again
    # and also avoid getting wrong answers
    visited.append( currNode )
    traverse_up( currNode.leftChild, inpNode, dist, count=count, outNodes=outNodes, visited=visited )
    traverse_up( currNode.rightChild, inpNode, dist, count=count, outNodes=outNodes, visited=visited )
    return outNodes

# Build the tree for testing
#      1 
#     / \ 
#   2    3 
#  /\   /\
# 4 5  6 7
testTree = build_test_tree()
print "#### nodes that the given distance apart from the input node ####"
print "nodes that given dist away from", testTree.leftChild.leftChild.key, "-->",
print get_dist_nodes( testTree.leftChild.leftChild, testTree, 4 )
print "#### nodes that the given distance apart from the input node ####"
```

```python
class MaxSum(object):
    
    def __init__( self, tree ):
        self.tree = tree
    
    def max_sum_leaves( self ):
        # we'll use a helper function to calculate/update maxSum
        self.maxSum=None
        self.max_sum_leaves_helper( self.tree )
        return self.maxSum

    def max_sum_leaves_helper( self, root, currMax=None ):
        # The basic layout of this problem is similar
        # to some of the previous one's. We just expand
        # on the root to leaf path sum problem. Here 
        # we'll check max sum in left subtree, then in right subtree
        # then have a third value with leftSum + rightSum + root value.
        # Compare the values with prev max and recurse so on.
        if root is None :
            return 0
        if currMax is None:
            currMax = root.key
        # find the max sums in the right and left subtrees
        leftSum = self.max_sum_leaves_helper( root.leftChild, currMax=currMax )
        rightSum = self.max_sum_leaves_helper( root.rightChild, currMax=currMax )
        # Check if the left subtree has the max, or the right subtree has the max
        # or if it is the path through the left subtree, right subtree and root
        currMax = max( (leftSum+rightSum+root.key), max(leftSum, rightSum) )
        if self.maxSum is None:
            self.maxSum = currMax
        else:
            if currMax > maxSum:
                self.maxSum = currMax
        # Now return the maximum root to leaf sum
        return max(leftSum, rightSum) + root.key

testTree = build_test_tree()
print "#### Max leaf to leaf sum path ####"
mObj = MaxSum( testTree )
print mObj.max_sum_leaves()
print "#### Max leaf to leaf sum path ####"
```

> Find the lowest common ancestor in a BST.

> Example:
>> 20 

>>/\ 

>>8      22

>>/ \  

>>4    12

>>     /\

>>     10 14

>> LCA of 10,14 is 12. LCA of 8 and 14 is 8.

> Source : http://www.geeksforgeeks.org/lowest-common-ancestor-in-a-binary-search-tree/

```python
def get_common_ancestor( root, node1, node2 ):
    # We'll make use of BST property here -> the root node
    # is always greater than the left child and less than right.
    # We'll iteratively traverse the tree (ofcourse from root to leaf)
    # and keep track of nodes whose values are between node1 and node2
    # (including node1 and node2).
    # If while iterating through the tree we encounter a node, whose
    # value is less than node1 and node2, then the nodes are on the right
    # and vice versa.
    lcaNode = None
    while root is not None:
        # If both nodes are smaller than root, then move left
        if ( (root.key > node1.key) and (root.key > node2.key) ):
            root = root.leftChild
            lcaNode = root.key
        # If both nodes are greater than root, then move right
        elif ( (root.key < node1.key) and (root.key < node2.key) ):
            root = root.rightChild
            lcaNode = root.key
        # Otherwise we're done
        else : break
    # NOTE : HERE we assume both nodes are in the BST.
    return lcaNode
    
    
# Build a tree for testing purposes
# with the structure:
#      20 
#     / \ 
#   8    22 
#  / \   
# 4  12  
#   / \
#  10 14
testTree = Tree(20)
testTree.insert_left_child(8)
testTree.insert_right_child(22)
testTree.leftChild.insert_left_child(4)
testTree.leftChild.insert_right_child(12)
testTree.leftChild.rightChild.insert_left_child(10)
testTree.leftChild.rightChild.insert_right_child(14)
# We'll use two test cases, one with 10, 14 as nodes
case1Node1 = testTree.leftChild.rightChild.leftChild
case1Node2 = testTree.leftChild.rightChild.rightChild
# other with 8, 14 as nodes
case2Node1 = testTree.leftChild
case2Node2 = testTree.leftChild.rightChild.rightChild
print "least common ancestor of 10, 14-->",
print get_common_ancestor( testTree, case1Node1, case1Node2 )
print "least common ancestor of 8, 14-->",
print get_common_ancestor( testTree, case2Node1, case2Node2 )
```
